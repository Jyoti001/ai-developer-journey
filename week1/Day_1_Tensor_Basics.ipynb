## Tensor - squeeze
#### squueze will only remove the dimensions of size 1. 
#### If we specifically describe the dimention, then only if the dimension is 1 then it will be squuezed else not

```python
shape = (1,3,1,5)
```

```python
t1 = torch.rand(shape)
```

```python
t1.shape
```

```python
t1
```

```python
t2 = t1.squeeze()
```

```python
t2.shape
```

```python
x = torch.rand(shape)
```

```python
x.shape
```

```python
x.squeeze(0).shape
```

```python
x.squeeze(1).shape ## will not change because the first dimension is not 1
```

## Tensor - unsqueeze() - opposite of squueze, will add a dimension of 1 at a specific index

```python
x.shape
```

```python
x.unsqueeze(0).shape
```

```python
x.unsqueeze(4).shape
```
