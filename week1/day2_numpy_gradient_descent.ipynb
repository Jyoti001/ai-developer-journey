```python
import numpy as np
```

```python
# num_of_datapoints = 100
# simple equation : y = 2x + 3 + noise
# only 1 feature (x) and 1 label (y)m
```

```python
x = np.linspace(0,1000,100)
```

```python
b = np.ones(100) * 3
```

```python
x.shape
```

```python
b.shape
```

```python
c = np.random.rand(100) 
```

```python
y = 2*x + b + c
```

```python
y
```

```python
## Now making the predictions
```

```python
epochs = 20
learning_rate = 0.0001
w = 0
b = 0
epoch_list = []
error_list = []
for epoch in range(epochs):
    y_pred = w*x + b
    error = ((y_pred - y) ** 2)
    # print("w = {} and b = {}".format(w, b))
    print("epoch {0} : {1}".format(epoch, error.mean()))
    epoch_list.append(epoch)
    error_list.append(error.mean())
    w = w - (learning_rate * ((-2 * x) * (y - y_pred)))
    b = b - (learning_rate * (-2 * (y - y_pred)))

```

```python
import matplotlib.pyplot as plt
```

```python
len(epoch_list)
```

```python
plt.plot(epoch_list, error_list)
```
