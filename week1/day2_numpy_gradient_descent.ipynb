```python
import numpy as np
```

```python
# num_of_datapoints = 100
# simple equation : y = 2x + 3 + noise
# only 1 feature (x) and 1 label (y)m
```

```python
x = np.linspace(0,1,100)
```

```python
b_constant = 3
```

```python
x.shape
```

```python
noise = np.random.rand(100) 
```

```python
y = 2*x + b_constant + noise
```

```python
y
```

```python
## Now making the predictions
```

```python
epochs = 1000
learning_rate = 0.0001
weights = 0
bias = 0
epoch_list = []
error_list = []
for epoch in range(epochs):
    y_pred = weights * x + bias
    error = ((y_pred - y) ** 2).mean() # scaler value
    # print("w = {} and b = {}".format(w, b))
    print("epoch {0} : {1}".format(epoch, error))
    epoch_list.append(epoch)
    error_list.append(error)
    weights = weights - (learning_rate * np.sum(((-2 * x) * (y - y_pred))))
    bias = bias - (learning_rate * np.sum((-2 * (y - y_pred))))

```

```python
import matplotlib.pyplot as plt
```
