```python
import torch
```

```python
from torchvision.models import resnet18, ResNet18_Weights
```

```python
model = resnet18(weights=ResNet18_Weights.DEFAULT)
```

```python
model
```

```python
data = torch.rand(1,3,64,64)
```

```python
type(data)
```

```python
data
```

```python
labels = torch.rand(1,1000)
```

```python
data.shape
```

```python
labels.shape
```

```python
## Forward pass
```

```python
prediction = model(data)
```

```python
type(prediction)
```

```python
prediction.shape
```

```python
## Error calculation
```

```python
error = (prediction - labels).sum()
```

```python
## Backward pass
```

```python
error.backward()
```

```python
## Optimizer
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
optim
```

```python
optim.step()
```
